## Calculate ML Points

#### Hitters

```python
# Test all theta columns are in DF
Thetas = HittersThetas.index.tolist()
Test = HittersProjected_ZScores[Thetas]

Test.head()
```

```python
# Generate ML Points (Multiply every player's projected stats by the thetas and adds the intercept)
HittersProjected_ZScores['ML Points'] = Test.dot(HittersThetas) + HittersIntercept
```

```python
# Define the number of rostered players for each position in league times 1.2 (In TeamScoresAnalysis.xlsx)
HittersRosterCounts = {
    'C': 16,
    '1B': 29,
    '2B': 24,  
    '3B': 25,  
    'SS': 26,  
    'OF': 66, 
    'DH': 17
}

# Create a new empty column for VORP
HittersProjected_ZScores['VORP'] = 0.0

# Loop through each position and calculate VORP dynamically
for position, count in HittersRosterCounts.items():
    # Isolate the current position and sort by projected points
    HittersTemp = HittersProjected_ZScores[HittersProjected_ZScores['Position'] == position].sort_values('ML Points', ascending=False)
    
    # Make sure we actually have enough players projected at this position to find the replacement level
    if len(HittersTemp) >= count:
        # Get the projected points of the replacement level player (count - 1 because Python is 0-indexed)
        ReplacementPoints = HittersTemp.iloc[count - 1]['ML Points']
        
        # Subtract the baseline from everyone at this specific position
        HittersProjected_ZScores.loc[HittersProjected_ZScores['Position'] == position, 'VORP'] = HittersProjected_ZScores['ML Points'] - ReplacementPoints
        
        # Optional: Print out the baseline so you can verify it's working!
        print(f"{position} Replacement Level (Player #{count}): {ReplacementPoints:.2f} ML Points")
        
    else:
        print(f"Warning: Not enough players projected at {position} to find replacement level.")

# Finally, sort your entire Hitters Big Board by VORP!
HittersProjected_ZScores = HittersProjected_ZScores.sort_values('VORP', ascending=False)

HittersProjected_ZScores.head()
```

```python
# Output the Final Hitter Draft Board
# Filter out players below replacement level (VORP < 0) and sort
HittersDraftBoard = HittersProjected_ZScores[HittersProjected_ZScores['VORP'] > 0][['Player Name', 'Position', 'Team', 'ML Points', 'VORP', 'ADP']]
HittersDraftBoard = HittersDraftBoard.sort_values('VORP', ascending=False)

HittersDraftBoard.head(10)
```

## Create Sleeper Score

```python
# Join the three dataframes
BigBoard = pd.concat([HittersDraftBoard, StartersDraftBoard, RelieversDraftBoard], ignore_index=True)

# Create Overall ranking
BigBoard['VORP Rank'] = BigBoard['VORP'].rank(ascending=False).astype(int)

# Move VORP Rank to the first column
cols = ['VORP Rank'] + [col for col in BigBoard.columns if col != 'VORP Rank']
BigBoard = BigBoard[cols]

BigBoard.head()
```

```python
# Calculate the Market Edge (Sleeper Score)
# Positive number = Sleeper (Public is drafting them too late)
# Negative number = Bust (Public is drafting them too early)
BigBoard['Sleeper Score'] = BigBoard['ADP'] - BigBoard['VORP Rank']

BigBoard.head()
```

## Create Position Tiers (KMeans)

```python
# Initialize an empty list to store the processed dataframes for each position
all_results = []

# Iterate through the BigBoard, isolating one player position at a time
for pos, group in BigBoard.groupby('Position'):

    # Create a clean copy sorted by VORP (highest to lowest) to prep for clustering
    group = group.copy().sort_values('VORP', ascending=False).reset_index(drop=True)
    n = len(group) # Get the total number of players at this specific position

    # Dynamically set the number of draft tiers based on the positional depth
    # Shallow positions get fewer tiers; deep pools get up to 5
    if n <= 5:
        n_tiers = 2
    elif n <= 15:
        n_tiers = 3
    elif n <= 30:
        n_tiers = 4
    else:
        n_tiers = 5

    # Reshape VORP into a 2D array (required by Scikit-Learn) and run KMeans clustering
    vorp_vals = group['VORP'].values.reshape(-1, 1)
    kmeans = KMeans(n_clusters=n_tiers, random_state=42, n_init=10)
    labels = kmeans.fit_predict(vorp_vals) # Assign each player to a raw, un-ordered cluster label

    # KMeans assigns cluster labels arbitrarily (e.g., Tier 0 isn't necessarily the best).
    # This block finds the average VORP of each cluster (centers), sorts those averages 
    # from highest to lowest, and maps them sequentially so Tier 1 is always the highest VORP.
    centers = kmeans.cluster_centers_.flatten()
    rank_map = {old: new + 1 for new, old in enumerate(np.argsort(centers)[::-1])}
    group['Tier'] = [rank_map[l] for l in labels]

    # Save the newly tiered positional group to our master list
    all_results.append(group)

# Combine all positional dataframes back into one master Big Board
# Sort neatly by Position alphabetically, then by Tier, and finally by VORP within each tier
BigBoard = pd.concat(all_results).sort_values(
    ['Position', 'Tier', 'VORP'],
    ascending=[True, True, False]
).reset_index(drop=True)

# Move VORP Rank and Tier to the first two columns
cols = ['VORP Rank', 'Tier'] + [col for col in BigBoard.columns if col not in ['VORP Rank', 'Tier']]
BigBoard = BigBoard[cols]

BigBoard.head(20)
```

## Export to CSV

```python
# sort the dataframe by the rank
BigBoard = BigBoard.sort_values(by='VORP Rank').reset_index(drop=True)

# Export final product to csv
BigBoard.to_csv("BigBoard.csv", index=False)
```

## Import Packages

```python
import pandas as pd
```

## Load Data

```python
# Load ADP data
HittersADP = pd.read_csv("AverageDraftPositionData/AverageDraftPositionDataClean/HittersADP2026Clean.csv")
PitchersADP = pd.read_csv("AverageDraftPositionData/AverageDraftPositionDataClean/PitchersADP2026Clean.csv")

# Load with latin1 to safely read the special characters
HittersProj = pd.read_csv("ProjectedDataRAW/HittersProj2026RAW.csv", encoding="utf-8")
PitchersProj = pd.read_csv("ProjectedDataRAW/PitchersProj2026RAW.csv", encoding="utf-8")
```

```python
# Display top elements of dataframe
HittersProj.head()
```

```python
# Display top elements of dataframe
PitchersProj.head()
```

## Clean Data

#### Remove Accent Marks

```python
# This automatically turns "Acuña" into "Acuna" and "José" into "Jose"
HittersProj['Player Name'] = HittersProj['Player Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
PitchersProj['Player Name'] = PitchersProj['Player Name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

print("Accent marks successfully removed!")
```

#### Normalize Team Abbreviations

```python
TeamNames = {
    'ARI': 'Arizona Diamondbacks', 'ATL': 'Atlanta Braves', 'BAL': 'Baltimore Orioles',
    'BOS': 'Boston Red Sox', 'CHC': 'Chicago Cubs', 'CHW': 'Chicago White Sox', 'CWS': 'Chicago White Sox',
    'CIN': 'Cincinnati Reds', 'CLE': 'Cleveland Guardians', 'COL': 'Colorado Rockies',
    'DET': 'Detroit Tigers', 'HOU': 'Houston Astros', 'KCR': 'Kansas City Royals', 'KC': 'Kansas City Royals',
    'LAA': 'Los Angeles Angels', 'LAD': 'Los Angeles Dodgers', 'MIA': 'Miami Marlins',
    'MIL': 'Milwaukee Brewers', 'MIN': 'Minnesota Twins', 'NYM': 'New York Mets',
    'NYY': 'New York Yankees', 'ATH': 'Athletics', 'OAK': 'Athletics', 'PHI': 'Philadelphia Phillies', 
    'PIT': 'Pittsburgh Pirates', 'SDP': 'San Diego Padres', 'SD': 'San Diego Padres',
    'SEA': 'Seattle Mariners', 'SFG': 'San Francisco Giants', 'SF': 'San Francisco Giants', 'STL': 'St. Louis Cardinals',
    'TBR': 'Tampa Bay Rays', 'TB': 'Tampa Bay Rays', 'TEX': 'Texas Rangers', 'TOR': 'Toronto Blue Jays',
    'WSN': 'Washington Nationals', 'WAS': 'Washington Nationals', 'WSH': 'Washington Nationals'
}

# Translate the abbreviations into full names
HittersProj['Team'] = HittersProj['Team'].map(TeamNames).fillna('Free Agent')
PitchersProj['Team'] = PitchersProj['Team'].map(TeamNames).fillna('Free Agent')
```

```python
# Display top elements of dataframe
HittersProj.head()
```

```python
# Display top elements of dataframe
PitchersProj.head()
```

```python
# Join projection stats with ADP data; 'inner' keeps only players found in both datasets.
Hitters = pd.merge(HittersProj, HittersADP, on=['Player Name', 'Team'], how='inner')

# Order column variables
HittersOrder = ['Player Name', 'Position', 'Team', 'Hits (H)', 'Runs Scored (R)', 'Total Bases (TB)', 'Runs Batted In (RBI)', 'Walks (BB)',
                'Intentional Walks (IBB)', 'Strikeouts (K)', 'Stolen Bases (SB)',	'ADP']
Hitters = Hitters[HittersOrder]

# Join projection stats with ADP data; 'inner' keeps only players found in both datasets.
Pitchers = pd.merge(PitchersProj, PitchersADP, on=['Player Name', 'Team'], how='inner')

# Order column variables
PitchersOrder = ['Player Name', 'Position', 'Team', 'Innings Pitched (IP)',	'Hits Allowed (H)',	'Earned Runs (ER)',	'Walks Issued (BB)', 
                 'Strikeouts (K)', 'Saves (SV)', 'Blown Saves (BS)', 'Holds (HD)', 'ADP']
Pitchers = Pitchers[PitchersOrder]
```

## Create CSV

```python
Hitters.to_csv("ProjectedDataClean/HittersProj2026Clean.csv", index=False)
Pitchers.to_csv("ProjectedDataClean/PitchersProj2026Clean.csv", index=False)

print("\nSaved as 'HittersProj2026Clean.csv' and 'PitchersProj2026Clean.csv'.")
```
